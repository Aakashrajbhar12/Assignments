### Assignment -- Recommendation System

Problem statement.

Build a recommender system by using cosine simillarties score.

Dataset --> bookRS.csv

```python
#Importing the Required Libraries
import pandas as pd
import numpy as np
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
#Loading the Dataset
books_df = pd.read_csv('C:/Users/Akaash/Downloads/bookRS.csv')
books_df.head()
```

Inference: Droping the unnamed columns as it is just index column

```python
#Dropping the Unnamed Column
books_df = books_df.iloc[:,1:]
books_df.head()
```

```python
#Checking the Shape of the Dataset
books_df.shape
```

```python
#number of unique users in the dataset
len(books_df['User.ID'].unique())
```

```python
#number of unique Books in the dataset
len(books_df['Book.Title'].unique())
```

Inference:

Total/Unqiue Users = 2182, Total/Unqiue Books = 9659

Here Product is Books and users are customer, so all the unique user should come in row side, and unique Books on the column side. So now we can use crosstab / Pivot table

After Using a Crosstab / Pivot Table we will get (2182,9659)--(rows,columns)

```python
#Creating a Pivot Table DataFrame
user_books_df = books_df.pivot_table(index='User.ID', columns='Book.Title', values='Book.Rating').reset_index(drop=True)
user_books_df
```

Inference: Inference: Now you can see we have (2182,9659) shape with proper index,as the previous index was reset using -->reset_index(drop=True)

Where drop = True drops original index

```python
#Replaceing index with userid
user_books_df.index = books_df['User.ID'].unique()
user_books_df
```

Inference: Replacing the index column with user's user id

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df
```

Inference: Filling NA Values with 0 as considering that the users haven't given rating to those books which they didn't Read.

#### Checking Similiarity --Cosine Based Similarity

```python
#Checking Similiarity by using Cosine based Similarity -- passing an array of values
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
user_sim
```

```python
#Store the results of Cosine-Base Similarity in a dataframe
user_sim_df = pd.DataFrame(user_sim)

#Set the index and column names to user ids 
user_sim_df.index = books_df['User.ID'].unique()
user_sim_df.columns = books_df['User.ID'].unique()
#Getting First 5 Row and columns
user_sim_df.iloc[0:5,0:5]
```

Inference: Created the matrix of Similiarity, 1 = Similiarity between Two Users, 0 = No Similiarity
    
Now as you can see similiarity between ourselves will be always 1 so droping them i.e diagonal element

```python
#Filling Diagonal to 0
np.fill_diagonal(user_sim, 0)
#Getting First 5 Row and columns
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:]
```

Inference: All these users have simliarity between them.

```python
#Checking Details of 162113 & 161453 Users to Recommend
books_df[(books_df['User.ID']==162113) | (books_df['User.ID']==161453)]
```

```python
#Creating DataFrame of Two User_id for Join Statement
user_1=books_df[(books_df['User.ID']==162113)]
user_2=books_df[(books_df['User.ID']==161453)]
print(user_1['Book.Title'])
print(user_2['Book.Title'])
```

```python
#Join Books Reads by both and then we can recommend the ones which have not been Read-- Joins like SQL
pd.merge(user_1,user_2,on='Book.Title',how='outer')
```
