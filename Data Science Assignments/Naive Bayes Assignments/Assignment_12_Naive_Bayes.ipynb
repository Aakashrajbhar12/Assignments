### Naive Bayes Assignment

#### Problem Statement
Prepare a classification model using Naive Bayes for salary data 

Dataset --> SalaryData_Test.csv, SalaryData_Train.csv

```python
#Importing the Required Library
import pandas as pd
import numpy as np
from sklearn import preprocessing

#Naive Bayes Library
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

```python
# loading dataset
dataset = pd.read_csv('C:/Users/Akaash/Downloads/SalaryData_Train.csv')
dataset.head()
```

```python
#Checking the Dataset for Na Values & Dtypes of Column
dataset.info()
```

Inference: There are No NA Values Present,  Also Datasets Has X as well as the Y Variable Categorial, So for Y variable Using Label Encoder then for X Variable will Use Dummies

```python
#Complete SalaryData dataset - applying lable encoder to size_category column - (Y Variable is Categorial)
label_encoder = preprocessing.LabelEncoder()
dataset['Salary'] = label_encoder.fit_transform(dataset['Salary']) 
dataset.head()
```

Inference: Label Encoder is Applied on Y Variable So 0 = "<=50" Class , 1 = ">50" Class

```python
#Dummies for X Variable
dataset = pd.get_dummies(dataset)
dataset.head()
```

Inference: Now the Dataset is Complete, Lets Separate X & Y Variable 

```python
# Dividing our data into X and Y variables 
X = pd.concat([dataset.iloc[:,0:5],dataset.iloc[:,6:]],axis = 1)
Y = dataset.iloc[:,5]
```

Inference: There are 102 Input Variable & 1 Class Variable

#### Naive Bayes Train Model

```python
# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(X,Y)
train_pred = classifier_mb.predict(X)
accuracy_train = np.mean(train_pred==Y)

print("Train Accuracy",accuracy_train)
```

Inference: Train Accuracy is 77.29% by Multinomial Naive Bayes 

```python
# Gaussian Naive Bayes 
classifier_gb = GB()
classifier_gb.fit(X,Y) 
train_pred_g = classifier_gb.predict(X)
accuracy_train_g = np.mean(train_pred_g==Y)

print("Train Accuracy",accuracy_train_g)
```

Inference: Train Accuracy is 80.3% by Gaussian Naive Bayes 

**Testing on the Test Data**

```python
# loading dataset
dataset = pd.read_csv('C:/Users/Akaash/Downloads/SalaryData_Test.csv')
#Complete Forestfires dataset - applying lable encoder to size_category column - (Y Variable is Categorial)
label_encoder = preprocessing.LabelEncoder()
dataset['Salary'] = label_encoder.fit_transform(dataset['Salary']) 
#Dummies for X Variable
dataset = pd.get_dummies(dataset)
# Dividing our data into X and Y variables
X_Test = pd.concat([dataset.iloc[:,0:5],dataset.iloc[:,6:]],axis = 1)
Y_Test = dataset.iloc[:,5]
```

```python
# Multinomial Naive Bayes (Predicting)
test_pred_m = classifier_mb.predict(X_Test)
accuracy_test_m = np.mean(test_pred_m==Y_Test)
print("Test Accuracy",accuracy_test_m)
```

Inference: Test Accuracy is 77.49% by Multinomial Naive Bayes 

```python
#Gaussian Naive Bayes (Predicting)
test_pred_g = classifier_gb.predict(X_Test)
accuracy_test_g = np.mean(test_pred_g==Y_Test)
print("Test Accuracy",accuracy_test_g)
```

Inference: Test Accuracy is 80.29% by Gaussian Naive Bayes 

##### Inference: So the Gaussian Naive Bayes Model is best with Test Accuracy of 80.29%
